# 📊 Projeto: Análise de Vendas de uma Loja

<p>
Este projeto tem como objetivo realizar uma análise exploratória de um conjunto de dados contendo o histórico de vendas de uma loja. Para isso, serão utilizadas as bibliotecas <strong>Pandas</strong> e <strong>Matplotlib</strong>, permitindo a importação, o tratamento, a organização e a visualização dos dados. Ao longo da análise, serão extraídas informações relevantes sobre o desempenho das vendas, auxiliando na identificação de padrões, tendências e indicadores importantes para a tomada de decisões.
</p>

<p>Preparando o ambiente</p>

```python
import matplotlib.pyplot as plt
import pandas as pd
dados = pd.read_csv('Material 26 - Vendas.csv', delimiter=';')
```

<p>1 - Quanto foi vendido por cidade? </p>

```python
quantidade = dados['QUANTIDADE'].astype(int)

preco = dados['PREÇO UNITÁRIO'].replace(',', '.').astype(float)

dados['Valor vendido'] = quantidade * preco

vendas_por_cidade = dados.groupby('CIDADE')['Valor vendido'].sum()

plt.plot(vendas_por_cidade.index, vendas_por_cidade.values)
plt.title('Valor vendido por cidade')
plt.xlabel('Cidade')
plt.ylabel('Valor vendido')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

<p>2 - Quais dias mais vendem? </p>

```python
dados['dia'] = dados['DATA E HORA COMPRA'].str[:2].astype(int)

valoresPorDia = round(dados.groupby('dia')['Valor vendido'].sum())

plt.plot(valoresPorDia.index, valoresPorDia.values)
plt.title('Dias que mais vendem')
plt.xticks(range(1, 32),rotation=45, ha='right')
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

<p> 3 - Qual é a quantidade vendida por hora no dia? </p> 

```python
dados['hora'] = dados['DATA E HORA COMPRA'].str[11:13].astype(int)

vendasHora = dados.groupby('hora')['QUANTIDADE'].sum()

plt.plot(vendasHora.index, vendasHora.values)
plt.title('Quantidade vendida por hora no dia')
plt.xlabel('Hora do dia')
plt.ylabel('Valor vendido')
plt.xticks(range(0, 24), rotation=70, ha='right')
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

<p>4 - Quantidade vendidas por produto? </p>

```python
dados['PREÇO UNITÁRIO'] = dados['PREÇO UNITÁRIO'].replace(',', '.').astype(float)
produtosSomas = dados.groupby('PRODUTO')['PREÇO UNITÁRIO'].sum()

plt.figure(figsize=(12, 5))
plt.plot(produtosSomas.index, produtosSomas.values, marker='o')
plt.title('Quantidade vendidas por produto')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=70, ha='right')
plt.tight_layout()

plt.show()
```

<p>5 - Quantos clientes tem ativo por cidade?</p>

```python
clientesPorCidade = dados.groupby('CIDADE')['CÓDIGO CLIENTE'].sum()

plt.plot(clientesPorCidade.index, clientesPorCidade.values, marker='o')
plt.title('Quantidade clientes por cidade')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=70, ha='right')
plt.tight_layout()

plt.show()
```
